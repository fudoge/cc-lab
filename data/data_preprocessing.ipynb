데이터 전처리

```python
# 엑셀 로드
import pandas as pd
df = pd.read_excel('restroom_data.xlsx')

# 필요한 컬럼만 추출
filtered_data = df[['화장실명', '전화번호', '개방시간상세', 'WGS84위도', 'WGS84경도']]
filtered_data
```

```python
# 결측값 제거
cleaned_data = filtered_data.dropna()
cleaned_data
```

```python
cleaned_data.columns = ['restroom_name', 'phone_number', 'open_time', 'latitude', 'longitude']
cleaned_data
```

```python
# csv로 저장
cleaned_data.to_csv("clean_restroom_data.csv", index=False)
```
